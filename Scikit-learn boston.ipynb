Plotting cross-calidated predictions
http://scikit-learn.org/stable/auto_examples/plot_cv_predict.html#example-plot-cv-predict-py

```python
from sklearn import datasets, cross_validation, linear_model
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
lr = linear_model.LinearRegression()
boston = datasets.load_boston()
predicted = cross_validation.cross_val_predict(lr, boston.data, boston.target, cv=10)
```

```python
fig, ax = plt.subplots()
ax.scatter(boston.target, predicted)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
```

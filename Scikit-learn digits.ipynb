# Scikit-learn tutorial

This script is mostly taken from the tutorial at http://scikit-learn.org/stable/tutorial/basic/tutorial.html#machine-learning-the-problem-setting

```python
import matplotlib.pyplot as plt
%matplotlib inline  

from sklearn import datasets, svm, metrics

digits = datasets.load_digits()
n_samples = len(digits.images)
```

```python
plt.figure(figsize=(20., 10.))
for index, (image, label) in enumerate(zip(digits.images[:40], digits.target[:40])):
    plt.subplot(4, 10, index + 1)
    plt.axis('off')
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title('Training: %i' % label)
```

```python
classifier = svm.SVC(gamma=0.001, C=100.)
classifier.fit(digits.data[:n_samples // 2], digits.target[:n_samples // 2])
```

```python
expected = digits.target[n_samples // 2:]
predicted = classifier.predict(digits.data[n_samples // 2:])

print("Classification report for classifier %s:\n%s\n"
      % (classifier, metrics.classification_report(expected, predicted)))
print("Confusion matrix:\n%s" % metrics.confusion_matrix(expected, predicted))
```

```python
mispredicted = [(pred, exp, data) for pred, exp, data in zip(predicted, expected, digits.images[n_samples // 2:]) if pred != exp]
plt.figure(figsize=(20., 7.5))
for index, (pred, exp, image) in enumerate(mispredicted):
    plt.subplot(3, 10, index + 1)
    plt.axis('off')
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title('real: %i,\n predicted: %i' % (exp, pred))
```
